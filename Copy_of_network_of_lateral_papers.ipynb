```python
import numpy as np
from pprint import pprint

def clean_authors(auth, replace={"{'e}": 'é', "{\`e}": 'è', "{\`o}": 'ò', '{"\i}': "ï"}):
  "replace LaTeX accents by ascii ones"
  txt = auth
  for k, v in replace.items():
    txt = txt.replace(k, v)
  return txt

def clean_dict(dic):
  "remove nan keys"
  return {k: v for k, v in dic.items() if isinstance(v, str)} #np.isfinite(v)}

# get the paper from the filename
def paper_from_filename(filename, bib_data_df=bib_df):
  candidates_df = bib_data_df.set_index('filename').loc[filename, :]
  if (candidates_df.shape[0]==1) or len(candidates_df.shape)==1:
    return clean_dict(candidates_df.to_dict())
  ind = candidates_df.reset_index()['ratio'].idxmax()
  return clean_dict(candidates_df.reset_index().loc[ind,:].to_dict())

found_bib = paper_from_filename('DEEP LEARNING FOR MEAN FIELD GAMES AND MEAN FIELD CONTROL WITH APPLICATIONS TO FINANCE - 2107.04568.pdf')
pprint(found_bib)
found_bib = paper_from_filename('Large Banking Systems with Default and Recovery: A Mean Field Game Model - 2001.10206.pdf')
pprint(found_bib)
```

### Loop to create the connections

```python
for concept, row in papers2concepts_df.groupby('Concept'):
  print(f"{len(row)} papers connected by {concept}")
  papers_connections_df = pandas.DataFrame(np.full((nbe_papers, nbe_papers), False), index=paper_names, columns=paper_names)
  papers_connection_ind = pandas.Series(paper_ind, index=paper_names)
  # really connected by this concept
  doc_names = row.Document.values
  # remove outside doncept papers
  to_remove = papers_connections_df.index[~papers_connections_df.index.isin(doc_names)]
  papers_connections_df.drop(index=to_remove, columns=to_remove, inplace=True)
  papers_connection_ind.drop(index=to_remove, inplace=True)
  # connect
  papers_connections_df.loc[doc_names, doc_names] = True
  # No diagonal
  for i in range(papers_connections_df.shape[0]):
    papers_connections_df.iloc[i,i] = False

  papers_connections_dic[concept] = AdjacencyIndex(adj=papers_connections_df, ind=papers_connection_ind)
```
